```python
def remove_leading_space(content):
    text = ""
    for i in range(1, len(content)):
        
        prev = content[i - 1]
        if prev == f"\n":
            continue
        else:
            text += content[i]
    text = content[0] + text
    return text
```

```python
col = df["Prompt"].dropna()
df["Prompt"] = col.apply(remove_leading_space)
```
