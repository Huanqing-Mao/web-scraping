```python
import os
import pandas as pd
```

```python
os.listdir()
```

```python
dataframe = []
for filename in os.listdir():
    if filename.endswith(".csv"):
        df = pd.read_csv(filename)
        dataframe.append(df)

combined_df = pd.concat(dataframe, ignore_index=True)
combined_df
```

```python
combined_df.to_csv("combined.csv", index=False)
```
