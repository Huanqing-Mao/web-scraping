```python
import os
import pandas as pd
```

```python
os.listdir("../../output/")
```

```python
dataframe = []
output_dir = "../../output/"
for filename in os.listdir(output_dir):
    if filename.endswith(".csv"):
        df = pd.read_csv(os.path.join(output_dir, filename))
        dataframe.append(df)

combined_df = pd.concat(dataframe, ignore_index=True)
combined_df
```

```python
combined_df.to_csv("../../output/combined.csv", index=False)
```
